```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
space_low = 0.0
space_high = 10.0
def get_toy_data(size=10):
    x = np.arange(space_low, space_high + (space_high - space_low)/size, (space_high - space_low)/size)
    y =np.random.uniform(space_low, space_high, (len(x)))
    return x,y
    
```

```python
data = get_toy_data(10)
x, y = data
plt.scatter(x, y)
```

```python
model_no_reg = make_pipeline(PolynomialFeatures(len(x)), LinearRegression())
model_no_reg.fit(np.reshape(x, (-1,1)), np.array(y))
space = np.arange(space_low,space_high, 0.01)
predictions_no_reg = model_no_reg.predict(np.reshape(space, (-1,1)))
```

```python
model_ridge = make_pipeline(PolynomialFeatures(len(x)), Ridge(fit_intercept=True))
model_ridge.fit(np.reshape(x, (-1,1)), np.array(y))
predictions_ridge= model_ridge.predict(np.reshape(space, (-1,1)))
```

```python
plt.plot(space, predictions_no_reg, label='No regularization')
plt.plot(space, predictions_ridge, label='Ridge regression')
plt.scatter(x, y, label="original points", c='g')
plt.legend()
```
